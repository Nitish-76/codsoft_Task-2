# 3. Data pre processing

```python
import re
import string
exclude = string.punctuation
```

```python
def remove_punc1(text):
    return text.translate(str.maketrans('','',exclude))
```

```python
# Translating all the words to lower-case for the model to take it in easily.
imdb['Genre'] = imdb['Genre'].str.lower()
imdb['Director'] = imdb['Director'].str.lower()
imdb['Actor 1'] = imdb['Actor 1'].str.lower()
imdb['Actor 2'] = imdb['Actor 2'].str.lower()
imdb['Actor 3'] = imdb['Actor 3'].str.lower()
# Removing most of the punctuation in the columns
imdb['Genre'] = imdb['Genre'].apply(remove_punc1)
imdb['Actor 1'] = imdb['Actor 1'].apply(remove_punc1)
imdb['Actor 2'] = imdb['Actor 2'].apply(remove_punc1)
imdb['Actor 3'] = imdb['Actor 3'].apply(remove_punc1)
imdb['Year'] = imdb['Year'].apply(remove_punc1)
imdb.dtypes
```

```python
# Removing the spaces between actors and directoer's name for tokenizing it better.
imdb['Actor 1'] = [i.replace(" ", "") for i in imdb['Actor 1']]
imdb['Actor 2'] = [i.replace(" ", "") for i in imdb['Actor 2']]
imdb['Actor 3'] = [i.replace(" ", "") for i in imdb['Actor 3']]
imdb['Director'] = [i.replace(" ", "") for i in imdb['Director']]
```

```python
# For Genre 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
bv = CountVectorizer(ngram_range =(1,1))
dv = CountVectorizer(ngram_range =(1,1))
```

```python
# Printing all the genres present in the dataset
Bag_of_genres = cv.fit_transform(imdb['Genre'])
# print vocabulary
print(cv.vocabulary_)
```

```python
# Vector of the genre of the particular row
print(Bag_of_genres[5].toarray())
```

```python
imdb['Actors'] = imdb['Actor 1']+ ' ' + imdb['Actor 2'] + ' ' + imdb['Actor 3']
imdb.head()
```

```python
# No. of actors present along with assigning them a unique value.
Unigrams_of_actors = bv.fit_transform(imdb['Actors'])
print(len(bv.vocabulary_))
```

```python
# No. of directors present in the dataset and assinging them a unique value.
Unigrams_of_direc = dv.fit_transform(imdb['Director'])
print(dv.vocabulary_)
```

# Observations
Therefore the rating of the movie depends on the year, director, genre and actors.
Hence it does not depend on votes or duration.  

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
```

```python
imdb['Actors'] = tfidf.fit_transform(imdb['Actors']).toarray()
```

```python
imdb['Genre'] = tfidf.fit_transform(imdb['Genre']).toarray()
```

```python
imdb['Director'] = tfidf.fit_transform(imdb['Director']).toarray()
```

```python
imdb.head()
```

```python
X = imdb.drop(columns= ['Name','Votes','Rating','Duration','Actor 1','Actor 2','Actor 3'])
Y = imdb['Rating']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
X_train,X_test,Y_train,Y_test = train_test_split( X,Y, test_size =0.2 ,random_state=3)
print(X.shape,X_test.shape,X_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_std = scaler.transform(X_train)
```

```python
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std.std())
print(X_test_std.std())
```

```python
regressor = RandomForestRegressor(n_estimators=100, random_state = 0)

regressor.fit(X_train_std, Y_train)

Giver2 = regressor.predict(X_test_std)

print(Giver2)
print('The mean Squared error:', mean_squared_error(Y_test, Giver2))
print('The mean Absolute error:', mean_absolute_error(Y_test, Giver2))
print('R2 Score of this model is:', r2_score(Y_test, Giver2))
print('Result by using RandomForestRegressor is this.')
```

```python
Model= LinearRegression()

Model.fit(X_train_std, Y_train)

Giver1 = Model.predict(X_test_std)

print(Giver1)
print('The mean Squared error:', mean_squared_error(Y_test, Giver1))
print('The mean Absolute error:', mean_absolute_error(Y_test, Giver1))
print('R2 Score of this model is:', r2_score(Y_test, Giver1))
print('Result by using RandomForestRegressor is this.')
```
